```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python

df = sns.load_dataset('diamonds')
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python

 
#code here
import seaborn as sns
import matplotlib.pyplot as plt

sns.set({'figure.figsize':(13,13)})

palette = "ch:r=-.2,d=.3_r"

sns.scatterplot(data=df, x='kolom_x', y='kolom_y', palette=palette)

plt.title('Scatterplot dengan Palette Kustom')
plt.xlabel('Label Sumbu X')
plt.ylabel('Label Sumbu Y')

plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")

sns.set({'figure.figsize':(10, 6)})

sns.boxplot(data=data, x='cut', y='price', palette="Set3")

plt.title('Boxplot by Cut for Price')
plt.xlabel('Cut')
plt.ylabel('Price')

plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python

#code here
import seaborn as sns
import matplotlib.pyplot as plt

df = df.sample(n=300, random_state=123)

markers = ['p', 'o', 's', 'D', 'X']

sns.pairplot(df, vars=['carat', 'depth', 'table', 'price'], markers=markers, diag_kind='kde')

plt.show()
```

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white", color_codes=True)
sns.jointplot(x="carat", y="price", data=df, kind="scatter", height=7, marker='x', color='b')
plt.subplots_adjust(top=0.9)
plt.suptitle("Scatter Joint Plot antara Carat dan Price", fontsize=16)


sns.set(style="white", color_codes=True)
sns.jointplot(x="carat", y="price", data=df, kind="kde", height=7, color='g')
plt.subplots_adjust(top=0.9)
plt.suptitle("KDE Joint Plot antara Carat dan Price", fontsize=16)

plt.show()
```
